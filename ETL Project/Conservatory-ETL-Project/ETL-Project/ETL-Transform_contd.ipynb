```python
# import dependencies
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
# from sqlalchemy import create_engine
# from config import DATABASE_URI
```

```python
# Inspect the files to view the table headers

# netflix_movies_revenue

NetflixRevenueFinal = pd.read_csv('netflix_movies_revenue.csv')
NetflixRevenueFinal.tail()
```

```python
# all_budget_revenue_final.csv
AllBudgetRevenueFinal = pd.read_csv('all_budget_revenue_final.csv')
AllBudgetRevenueFinal.tail()
```

```python
# MoviesFinalBudget.csv
MoviesFinalBudget = pd.read_csv('MoviesFinalBudget.csv')
MoviesFinalBudget.head()
```

```python
# Rename Movie column to title to enable merge
MoviesFinalBudgetRenamed = MoviesFinalBudget.rename(columns = {'Movie': 'title'})
MoviesFinalBudgetRenamed.head()
```

```python
# Merge AllBudgetRevenueFinal and MoviesFinalBudgetRenamed
MoviesBudget = pd.merge(AllBudgetRevenueFinal, MoviesFinalBudgetRenamed, on = 'title')
```

```python
MoviesBudget.head()
```

```python
# Check for duplicates
MoviesBudget= MoviesBudget[MoviesBudget.duplicated(subset=['id'])]
```

```python
# Drop duplicates
MoviesBudget = MoviesBudget.drop_duplicates(subset=['id'], keep='last')
```

```python
runtimes = MoviesBudget.loc[MoviesBudget['runtime'].isna()]
runtimes
```

```python
# Fill nan value with average movie time
MoviesBudget['runtime'] = MoviesBudget['runtime'].fillna(97)
```

```python
# Export as csv FinalMoviesBudget.csv
MoviesBudget.to_csv('FinalMoviesBudget.csv', index = False)
```
