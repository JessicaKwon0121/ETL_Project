### Web Scraping - The Numbers Movies Budget

```python
from splinter import Browser
from bs4 import BeautifulSoup
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import numpy as np
from time import sleep
from random import randint
```

```python
executable_path = {'executable_path': ChromeDriverManager().install()}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
# The budget
# Extract first 6000 movies from The Numbers website 
# Use splinter and beautiful soup to extract data from the first 60 pages
# Save data csv
```

```python
# Range of first 60 pages to loop through
urls = np.arange(1, 6001, 100)
```

```python
# create empty list to hold the tables
movies = []
for url in urls:
    
    print('Extracting data')
    print('.................')
    
    # pages to loop through
    page = 'https://www.the-numbers.com/movie/budgets/all' + '/' + str(url)
    
    # Browser instance t visit pages
    browser.visit(page)
    
    # Extract htlp of pages
    html = browser.html
    
    # Create a Soup instance
    soup = BeautifulSoup(html, 'html.parser')
    
    # Extract the tables and allow minimum 2 seconds wait time to sleep
    data = soup.find_all('table')
    sleep(randint(2,30))
    
    # Create a dataframe from the tables
    film = pd.read_html(str(data))[0]
    
    # Append the dataframe to the movies list
    movies.append(film)
   
```

```python
# Get complete dataframe of all the dataframes in movies
# create an empty dataframe
df = pd.DataFrame()

# loop through the movies list of dataframes
for movie in movies:
    
    # append the movies in the the empty dataframe df
    df = df.append(movie)
```

```python
# Reset index of df
df_movie = df.reset_index()
```

```python
# Inspect df_movie
df_movie.tail()
```

```python
# export movie_df as csv file df_6000_movie.csv
df_movie.to_csv('df_6000_movie.csv', index = False)
```

```python
browser.quit()
```
